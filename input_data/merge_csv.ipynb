```python
import pandas as pd
from path import Path
```

```python
def csv_to_df(stock_csv, sentiment_results_csv):
    file_path1 = Path(stock_csv)
    file_path2 = Path(sentiment_results_csv)
    stock_df = pd.read_csv(file_path1)
    sentiment_result = pd.read_csv(file_path2)
    
    return stock_df, sentiment_result
```

```python
csv_to_df("../stock_data/tsla.csv", "../tsla_sentiment.csv")
```

```python
tsla_stock_df, tsla_sent_df = csv_to_df("../stock_data/tsla.csv", "../tsla_sentiment.csv")
```

```python
tsla_stock_df['Date'] = pd.to_datetime(tsla_stock_df.Date, format='%Y-%m-%d')
```

```python
tsla_sent_df.dropna()
```

```python
tsla_sent_df['Date'] = pd.to_datetime(tsla_sent_df.created_at, format='%Y-%m-%d')
sum_tsla_sent = tsla_sent_df.groupby(tsla_sent_df['Date'].dt.date).mean()
```

```python
print(len(tsla_stock_df))
print(len(sum_tsla_sent))
```

```python
sum_tsla_sent.head()
```

```python
tsla_stock_df = tsla_stock_df.set_index("Date")
```

```python
final_df = pd.concat([tsla_stock_df, sum_tsla_sent], axis=1)
```

```python
final_df.to_csv("ML_tsla_input_df", encoding='utf-8', index=False)
```
