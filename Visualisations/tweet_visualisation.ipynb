```python
import pandas as pd
from path import Path
from sklearn import metrics
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def csv_to_df(input_csv):
    file_path = Path(input_csv)
    input_df = pd.read_csv(file_path)
    
    return input_df
```

```python
aapl_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/aapl_sent_results.csv")
btc_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/btc_sent_results.csv")
pfe_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/pfe_sent_results.csv")
nflx_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/nflx_sent_results.csv")
msft_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/msft_sent_results.csv")
tsla_tweets = csv_to_df("../sentiment_analysis/sentiment_analysis/tsla_sent_results.csv")
```

```python
#plots the top 20 used words

plt.figure(figsize=(10,5))
sns.barplot(aapl_tweets["tweet"], alpha=0.8)
plt.title('Top Words Overall')
plt.xlabel('Count of words', fontsize=12)
plt.ylabel('Word from Tweet', fontsize=12)
plt.show()
```

```python
def plot_polarity(df, name):
    fig, ax = plt.subplots(figsize=(6, 4))

    # Plot histogram of the polarity values
    df["compound"].hist(bins=[-1, -0.75, -0.5, -0.25, 0.25, 0.5, 0.75, 1],
                 ax=ax,
                 color="purple")

    plt.title("Polarity from Tweets on " + name)
    plt.show()
```

```python
plot_polarity(aapl_tweets, "Apple")
```

```python
plot_polarity(tsla_tweets, "Tesla")
```

```python
plot_polarity(pfe_tweets, "Pfizer")
```

```python
plot_polarity(nflx_tweets, "Netflix")
```

```python
plot_polarity(msft_tweets, "Microsoft")
```

```python
plot_polarity(btc_tweets, "Bitcoin")
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color='white',
        max_words=200,
        max_font_size=40, 
        scale=3,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data))

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud)
    plt.show()

show_wordcloud(aapl_tweets["tweet"])
```
